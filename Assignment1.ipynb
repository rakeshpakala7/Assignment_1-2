# Rename all columns in uppercase.

```python

uber_data = uber_data.rename(columns={'START_DATE*': 'START_DATE','END_DATE*': 'END_DATE','START*': 'START','STOP*':'STOP','MILES*':'MILES','PURPOSE*':'PURPOSE'})
uber_data
```

# Assignment 1.2

# Display rides where CATEGORY == 'Business'.

```python
b_ride=uber_data[uber_data['CATEGORY*']=='Business']
b_ride
```

# Show top 5 rides with the longest distance (MILES).

```python
top=uber_data.sort_values(by='MILES',ascending=False).head(5)
top
```

# Replace all missing PURPOSE values with "Not Specified".

```python
uber_data['PURPOSE'] = uber_data['PURPOSE'].fillna('Not Specified')
uber_data['PURPOSE']
```

```python
uber_data['START_DATE'] = pd.to_datetime(uber_data['START_DATE'])
uber_data['END_DATE'] = pd.to_datetime(uber_data['END_DATE'])
uber_data['TRIP_DURATION'] = uber_data['END_DATE'] - uber_data['START_DATE']
uber_data['TRIP_DURATION']
```

# Sort trips by distance in descending order.

```python
ub = uber_data.sort_values('MILES', ascending=False)
ub
```
